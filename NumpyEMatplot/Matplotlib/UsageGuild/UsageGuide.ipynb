###### Usage Guide
This tutorial covers some basic usage patterns and best-pratices to help you started with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### a simple example

Matplotlib graphs your data on Figures (i.e. , windows, Jupyter widgets, etc.), each of which can contain one or more Axes (e.e., an area where points can be specified in terms of x-y coordinates, or theta-r in a polar plot, or x-y-z in a 3D plot, etc.) The simplist way of creating a figure with an axes is uding pyplot.subplots. We can then use Axes.plot to draw some data  on the axes:



```python
fig, ax = plt.subplots() # create a figure containing a simpgle axis,
ax.plot([1,2,3,4],[1,4,2,3]) # plot some data on the axes.
```

Many other plotting libraries or languages do not require you to explicity create an axes. For example, in MATLAB, one can just do

```python
plot([1, 2, 3, 4],[1, 4, 2, 3])   #%MATLAB plot.
```

and get the desire graph.

in fact, you can do the same in Matplotlib.pypolot: for each Axes graphing method, there is a corresponding function in the matplot.pyplot module that performs that plot on the 'current' axes, creating that axes (and its parent figure) if they don't exist yet, So, the previus example can be written more shortly as


```python
plt.plot([1, 2, 3, 4],[1, 4, 2, 3]) # Matplotlib plot.
```

###### Parts of a figure

Now, let's have a deeper look at the components of a Matplot figure.

Figure

The Whole figure, the figure keeps track of all the child Axes,a smattering of 'special' artists (titles, figure legends, etc), and the canvas. (Don't worry too much about canvas, it is cricual as it is the object that actually does the drawing to get your plot, but as the user it's more or less invisible to you). A figure can contain any number of Axes, but will typically have at least one.

The easiest way to create a new figure is with pyplot



```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

it's convenience to create the axes together with the figure, but you can also add axes later on,
allowing for more complex axes layouts.

##### Axes
    This is what you think of as 'a plot' , it is the region of the image with the data space. A given figure can contain manu Axes, but a given Aves object  can only be in one figure. The axes contain two (or three axis em case of 3D) Axis objects (be aware of the difference between Axes and axis) which take care of the limits (the data limits cans also be controlled  via the axes.Axes.set.xlim()), an x-label (set via xlabel()),and a y-label (set via set_ylabel()).

    The axes class and its member functions are the primary entry point to working with the OO interfaces

#### Axis
    these are the number-line-like objects. they take care of setting the graph limits and generating the ticks (the marks on the axis) and ticklabels (strings labeling the ticks). the location of the ticks is determined by a locator object and tickelabel strings are formatted by a Formatter. The combination of the correct Locator and Formatter gives veryy fine control over the tick locations and labels.

#### Artist
    Basically, everything you can see on the figure is an artist (even the figure , axes, and axis objects). This includes Text Objects, Line2D objects, Patch objects ... (you get the idea). When the figure is rendered, all of the artists are drawn to the canvas. Most Artist are tied to an axis; Such an Artist cannot be shared by multiples axes, or moved from one to another.

    All of plotting functions expect NumPy.array or numpu.ma.masked_array as input. Classes that are 'array-like' suck as pandas data objects and numpy.matrix may or may not work as intended. it is best to convert these to numpy.array objects prior to plotting.

The object-oriented interface and the pyplot and the pyplot interface

As noted above, there are sessentially two ways to use Matplotlib:
    - Explicitly create figures and axes, and call methods on them (the "object-oriented(OO) style").
    - Rely on pyplot to automatically create and manage the figures and axes, and pyplot functions for plotting.

So one can do (OO-style)

```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

or (pyplot-style)

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```
